```
import numpy as np
import pandas as pd 
```

```
abalone_data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/abalone_data/training_set_label.csv" )
# print(abalone_data)
dummies = pd.get_dummies(abalone_data.Sex)
dummies
```

```
merged = pd.concat([abalone_data,dummies],axis=1)
merged
```

```
final = merged.drop(['Sex','M'], axis = 1)
final
```

```
X = final.drop('Rings', axis = 1) 

y = final.Rings             # Output/Dependent variable
print(X)
print(y)
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.3, random_state = 42)

print(X_train)
print(y_train)
```

```
from sklearn.ensemble import RandomForestRegressor
rfc = RandomForestRegressor()
rfc.fit(X_train, y_train)
```

```
pred = rfc.predict(X_val)
rfc.score(X,y,None)
```

```
from sklearn.metrics import mean_squared_error
print('Root Mean Squared Error is: ', np.sqrt(mean_squared_error(y_val, pred))) 
```

```
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/abalone_data/testing_set_label.csv')
dum_test = pd.get_dummies(test_data.Sex)
dum_test
```

```
merged_test = pd.concat([test_data,dum_test],axis=1)
merged_test
```

```
final_test = merged_test.drop(['Sex','M'], axis = 1)
final_test
```

```
target = rfc.predict(final_test)
```

```
res = pd.DataFrame(target) # target are nothing but the final predictions of your model on input features of your new unseen test data
res.index = final_test.index # its important for comparison. Here "test_new" is your new test dataset
res.columns = ["prediction"]

# To download the csv file locally
from google.colab import files
res.to_csv('submission.csv')         
files.download('submission.csv')
```
